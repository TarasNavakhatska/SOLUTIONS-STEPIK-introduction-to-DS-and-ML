```python
iris = pd.read_csv('https://stepic.org/media/attachments/course/4852/iris.csv', index_col = 0)
iris
```

```python
# DISPLOT
sns.distplot (iris["sepal length"], color="blue")
sns.distplot (iris["sepal width"] , color="red")
sns.distplot (iris["petal length"], color="green")
sns.distplot (iris["petal width"] , color="black")
```

```python
#  
name = 'petal width'
iris[name].max() - iris[name].min()
```

```python
name = 'petal length'
iris[name].max() - iris[name].min()
```

```python
name = 'sepal width'
iris[name].max() - iris[name].min()
```

```python
name = 'sepal length'
iris[name].max() - iris[name].min()
```

```python
# KDEPLOT
sns.kdeplot (iris["sepal length"], color="blue")
sns.kdeplot (iris["sepal width"] , color="red")
sns.kdeplot (iris["petal length"], color="green")
sns.kdeplot (iris["petal width"] , color="black")
```

```python
#   
sns.distplot(iris["sepal width"] , color="red", kde = False)
```

```python
#     VIOLINPLOT
sns.violinplot (iris["sepal length"], color="blue")
sns.violinplot (iris["petal length"], color="green")
```

```python
# orient ='v'   90 
sns.violinplot (iris["sepal length"], color="blue", orient='v')
sns.violinplot (iris["petal length"], color="green", orient='v')
```
