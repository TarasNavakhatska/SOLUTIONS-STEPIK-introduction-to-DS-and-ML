# ВОДОРОСЛИ


```python
concentrations = pd.read_csv('https://stepic.org/media/attachments/course/4852/algae.csv')
concentrations
```

```python
# Найти среднюю концентрацию каждого из веществ в каждом из родов (колонка genus)!
# Для этого проведите группировку датафрэйма, сохранённого в переменной concentrations,
# примените метод, сохранив результат в переменной mean_concentrations.
```

```python
# 1 способ
mean_concentrations = concentrations.groupby('genus')\
.agg({'sucrose': 'mean', 'alanin': 'mean', 'citrate': 'mean', 'glucose': 'mean', 'oleic_acid': 'mean'})
mean_concentrations
```

```python
# 2 способ
mean_concentrations = concentrations.groupby('genus').mean()
mean_concentrations
```

```python
# 2 способ
concentrations .groupby('genus').describe()['alanin']
```

```python
# Сгруппируйте данные по переменной group
# Размах содержания сахарозы в бурых, зелёных и красных водорослях
# Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях
# Число видов бурых, зелёных и красных водорослей
```

```python
# Размах содержания сахарозы в бурых, зелёных и красных водорослях
concentrations.groupby ('group').describe() ['sucrose']
```

```python
# Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях
concentrations.groupby ('group').describe() ['citrate']
```

```python
# Число видов бурых, зелёных и красных водорослей
concentrations.groupby ('group').count()
```
