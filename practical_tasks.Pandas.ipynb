```python
# 1 способ
df.loc[df["n_users"] < 0, "n_users"] = \
   df.loc[df["n_users"] >= 0, "n_users"].median()
df
```

# Рассчитайте среднее значение переменной session_value для каждой группы (переменная group), в получившемся dataframe  переменная group не должна превратиться в индекс. Также переименуйте колонку со средним значением session_value в mean_session_value.

```python
# число наблюдений группе 'time'. COUNT применяется ко всем колонкам!!!!
df.groupby('time').count()
```

```python
# число наблюдений группе 'n_users
df.groupby('n_users').count()
```

```python
# GROUP + AGG ПРИМЕНЯЕТ ФУНКЦИЮ К НУЖНОЙ КОЛОНКЕ
df.groupby('group').agg({'n_users': 'count'})
```

```python
# GROUP не должна превратиться в индекс
mean_session_value_data = df.groupby('group', as_index = False).agg({'session_value': 'mean'})\
.rename (columns = {'session_value' : 'mean_session_value'})
mean_session_value_data
```
